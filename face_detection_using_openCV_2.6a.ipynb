```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#read an image
image = cv2.imread(r'C:\Users\Siva Reddy\Desktop\computer_vision\modi.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#load the cascade
#name of xml file is haarcascade_frontalface_default.xml
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')

faces_rect = haar_cascade.detectMultiScale(
    gray, scaleFactor=1.1, minNeighbors=9)

for (x, y, w, h) in faces_rect:
    print(f'x:{x}, y:{y}, w:{w}, h:{h}')
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 0, 255), thickness=2) #(x,y) is left top corner, (x+w, y+h) right bottom

gray_2_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(gray_2_color)
plt.show()
```

# Face detection from live video

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

video = cv2.VideoCapture(0)

#load the cascade
#name of xml file is haarcascade_frontalface_default.xml
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')

while True:
    tag, image = video.read()
    #print(f'Tag:{tag},Image:{image}')
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    faces_rect = haar_cascade.detectMultiScale(
    gray, scaleFactor=1.1, minNeighbors=9)

    for (x, y, w, h) in faces_rect:
        #print(f'x:{x}, y:{y}, w:{w}, h:{h}')
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), thickness=2) #(x,y) is left top corner, (x+w, y+h) right bottom

    cv2.imshow('image', image)
    
    k = cv2.waitKey(5) & 0xff
    if k == 27: #ESc=27
        break
video.release()
cv2.destroyAllWindows()
```

# Face detection from recorded video

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

video = cv2.VideoCapture(r'C:\Users\Siva Reddy\Desktop\computer_vision\video2.mp4')

#load the cascade
#name of xml file is haarcascade_frontalface_default.xml
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')

while True:
    tag, image = video.read()
    #print(f'Tag:{tag},Image:{image}')
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    faces_rect = haar_cascade.detectMultiScale(
    gray, scaleFactor=1.1, minNeighbors=9)

    for (x, y, w, h) in faces_rect:
        #print(f'x:{x}, y:{y}, w:{w}, h:{h}')
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), thickness=2) #(x,y) is left top corner, (x+w, y+h) right bottom

    cv2.imshow('video', image)
    k = cv2.waitKey(1) & 0xff
    if k == 27: #27=Esc
        break
video.release()
cv2.destroyAllWindows()
```
